```python
#Unit sample sequence

import numpy as np
import matplotlib.pyplot as plt

def unit_sample_sequence(start, end):
    n = np.arange(start, end + 1)
    unit_sample = np.zeros_like(n)
    unit_sample[n == 0] = 1  # Set δ[n] = 1 when n = 0
    return n, unit_sample

#The Parameters
start = -10
end = 10

#Generating unit sample sequence
n, unit_sample = unit_sample_sequence(start, end)

#The Plotting
plt.stem(n, unit_sample)
plt.title('Unit Sample Sequence')
plt.xlabel('n')
plt.ylabel('δ[n]')
plt.xticks(n)
plt.grid()
plt.show()

```

```python
#unit step signal

import numpy as np
import matplotlib.pyplot as plt

def unit_step_signal(start, end):
    n = np.arange(start, end + 1)
    unit_step = np.where(n < 0, 0, 1)  # Set u[n] = 1 for n >= 0
    return n, unit_step

#The Parameters
start = -10
end = 10

#Generating unit step signal
n, unit_step = unit_step_signal(start, end)

#The Plotting
plt.stem(n, unit_step, basefmt=" ")
plt.title('Unit Step Signal')
plt.xlabel('n')
plt.ylabel('u[n]')
plt.xticks(n)
plt.grid()
plt.show()
```

```python
#unit ramp signal

import numpy as np
import matplotlib.pyplot as plt

def unit_ramp_signal(start, end):
    n = np.arange(start, end + 1)
    unit_ramp = np.where(n < 0, 0, n)  # Set r[n] = n for n >= 0
    return n, unit_ramp

#The Parameters
start = -10
end = 10

#Generating unit ramp signal
n, unit_ramp = unit_ramp_signal(start, end)

#The Plotting
plt.stem(n, unit_ramp, basefmt=" ")
plt.title('Unit Ramp Signal')
plt.xlabel('n')
plt.ylabel('r[n]')
plt.xticks(n)
plt.grid()
plt.show()
```

```python
#exponential signal

import numpy as np
import matplotlib.pyplot as plt

def exponential_signal(start, end, a):
    n = np.arange(start, end + 1)
    exp_signal = np.exp(a * n)  # Generate exponential signal
    return n, exp_signal

#The Parameters
start = -10
end = 10
a = 0.5  # Change this value for different growth/decay rates

#Generating exponential signal
n, exp_signal = exponential_signal(start, end, a)

#The Plotting
plt.stem(n, exp_signal, basefmt=" ")
plt.title('Exponential Signal')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.xticks(n)
plt.grid()
plt.show()
```

```python
#Square wave signal

import numpy as np
import matplotlib.pyplot as plt

def square_wave_signal(start, end, frequency):
    n = np.arange(start, end + 1)
    # Generate square wave signal: 1 for positive half, -1 for negative half
    square_wave = 1 * (np.sin(2 * np.pi * frequency * n) >= 0)
    square_wave = 2 * square_wave - 1  # Convert to -1 and 1
    return n, square_wave

#The Parameters
start = 0
end = 10
frequency = 1  # Frequency of the square wave

#Generating square wave signal
n, square_wave = square_wave_signal(start, end, frequency)

#The Plotting
plt.stem(n, square_wave, basefmt=" ")
plt.title('Square Wave Signal')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.xticks(n)
plt.ylim(-1.5, 1.5)  # Set y-limits for better visualization
plt.grid()
plt.show()
```

```python
#Triangular signal

import numpy as np
import matplotlib.pyplot as plt

def triangular_signal(start, end, period):
    n = np.arange(start, end + 1)
    # Create a triangular wave
    triangular_wave = (2 / period) * (n % period)  # Ramp up
    triangular_wave[n % period > period / 2] = 2 - triangular_wave[n % period > period / 2]  # Ramp down
    return n, triangular_wave

#The Parameters
start = 0
end = 10
period = 5  # Length of one cycle

#Generating triangular signal
n, triangular_wave = triangular_signal(start, end, period)

#The Plotting
plt.stem(n, triangular_wave, basefmt=" ")
plt.title('Triangular Signal')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.xticks(n)
plt.ylim(-1, 2)  # Set y-limits for better visualization
plt.grid()
plt.show()
```
